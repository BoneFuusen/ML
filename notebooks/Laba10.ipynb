```python
nlp = spacy.load("en_core_web_sm")
```

```python
preprocessed_documents = []
for document in df.v2:
    tokens = nlp(document.lower())
    data_lematized = [word.lemma_ for word in tokens]
    preprocessed_documents.append(data_lematized)
df["v3"] = preprocessed_documents
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vectorised = CountVectorizer()
```

```python
stringed_words = []
for listik in df["v3"].values:
    string = ""
    for word in listik:
        string += word + " "
    stringed_words.append(string)
df["v4"] = stringed_words
```

## LDA.

```python
from time import time
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.datasets import fetch_20newsgroups
```

```python
lda = LatentDirichletAllocation(n_components=25, max_iter=20,
learning_method='online',
learning_offset=50.,
random_state=0).fit(tfidf)
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
```

```python
print_top_words(lda, tfidf_vectorizer.get_feature_names_out(), 10)
```

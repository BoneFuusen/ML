# 3. Обработка данных.

## 3.1. Изменение типов данных. 

```python
data["Diabetes_012"] = data["Diabetes_012"].astype(int)
data["HighBP"] = data["HighBP"].astype(int)
data["HighChol"] = data["HighChol"].astype(int)
data["CholCheck"] = data["CholCheck"].astype(int)
data["BMI"] = data["BMI"].astype(int)
data["Smoker"] = data["Smoker"].astype(int)
data["Stroke"] = data["Stroke"].astype(int)
data["HeartDiseaseorAttack"] = data["HeartDiseaseorAttack"].astype(int)
data["PhysActivity"] = data["PhysActivity"].astype(int)
data["Fruits"] = data["Fruits"].astype(int) 
data["Veggies"] = data["Veggies"].astype(int)
data["HvyAlcoholConsump"] = data["HvyAlcoholConsump"].astype(int)
data["AnyHealthcare"] = data["AnyHealthcare"].astype(int)
data["NoDocbcCost"] = data["NoDocbcCost"].astype(int)
data["GenHlth"] = data["GenHlth"].astype(int)
data["PhysHlth"] = data["PhysHlth"].astype(int)
data["DiffWalk"] = data["DiffWalk"].astype(int)
data["Sex"] = data["Sex"].astype(int)
data["Age"] = data["Age"].astype(int)
data["Education"] = data["Education"].astype(int)
data["Income"] =data["Income"].astype(int)
```

## 3.4. Выбросы.

```python
data.hist(figsize=(20,15))
```

```python
outlier = data[['BMI', 'MentHlth', 'PhysHlth']]
Q1 = outlier.quantile(0.25)
Q3 = outlier.quantile(0.75)
IQR = Q3-Q1
```

```python
data_filtered = outlier[~((outlier < (Q1 - 1.5 * IQR)) |(outlier > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
index_list = list(data_filtered.index.values)
```

```python
data_filtered = data[data.index.isin(index_list)]
data_filtered.head()
```

```python
filtered_outlier = data_filtered[['BMI', 'MentHlth', 'PhysHlth']]

for col in filtered_outlier:
    data_filtered.hist(col, bins=100)
```

```python
data_filtered.hist(figsize=(20,15))
```

<b>Мы избавились от выбросов. Выполним замещение изменений из data_filtered в data и приступим к EDA.<b>

```python
data = data_filtered
```

# 6. Задача регрессии.

## 6.1. Выделение целевого признака и предикторов.

Целевой признак - наличие(или отсутствие) диабета.

```python
y = data["Diabetes_012"]
X = data.drop(["Diabetes_012"], axis=1)
```

```python
y
```

```python
X
```

## 6.2. Разделение данных на обучающую и тестовую выборки.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

## 6.3. Линейная регрессия.

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
lr = LinearRegression().fit(X_train[['GenHlth']], y_train)
```

```python
lr.predict(X_test[['GenHlth']])
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from math import sqrt
```

```python
y_pred = lr.predict(X_test[['GenHlth']])
X_test_genhlth = X_test[['GenHlth']]
```

```python
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {sqrt(mean_squared_error(y_test, y_pred))}')
print(f'MAPE: {sqrt(mean_absolute_percentage_error(y_test, y_pred))}')
print(f'R^2: {lr.score(X_test_genhlth, y_test)}')
```

### 6.3.1. L1. Lasso.

```python
lasso = Lasso(alpha=0.001).fit(X_train[['GenHlth']], y_train)
y_pred = lasso.predict(X_test_genhlth)
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {sqrt(mean_squared_error(y_test, y_pred))}')
print(f'MAPE: {sqrt(mean_absolute_percentage_error(y_test, y_pred))}')
print(f'R^2: {lasso.score(X_test_genhlth, y_test)}')
lasso.coef_
```

### 6.3.2. L2. Ridge.

```python
ridge = Ridge(alpha=0.5).fit(X_train[['GenHlth']], y_train)
y_pred = lasso.predict(X_test_genhlth)
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {sqrt(mean_squared_error(y_test, y_pred))}')
print(f'MAPE: {sqrt(mean_absolute_percentage_error(y_test, y_pred))}')
print(f'R^2: {ridge.score(X_test_genhlth, y_test)}')
ridge.coef_
```

### 6.3.3. ElasticNet в линейной регрессии.

```python
from sklearn.linear_model import ElasticNet
```

```python
grid_model.best_params_
```

```python
print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
print(f'MSE: {mean_squared_error(y_test, y_pred)}')
print(f'RMSE: {sqrt(mean_squared_error(y_test, y_pred))}')
print(f'MAPE: {sqrt(mean_absolute_percentage_error(y_test, y_pred))}')
print(f'R^2: {grid_model.score(X_test_genhlth, y_test)}')
ridge.coef_
```

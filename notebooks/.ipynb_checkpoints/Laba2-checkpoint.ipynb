# 1. Инициализация проекта.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
```

```python
data = pd.read_csv('../data/DataSet2.csv')
```

# 2. Исследование исходных данных. 

```python
data.info()
```

```python
data.head(10)
```

##### Отметим основные проблемы:

1) Название столбца `Unnamed: 0`
2) Неправильные типы данных в столбцах: `Streaming Release Date`, `Original Release Date`, `IMDB Votes` и `Oscar Year`
3) Пропуски данных

# 3. Обработка данных. 

## 3.1. Исправление типов данных.

##### Oscar Year

```python
data['Oscar Year'].unique()
```

```python
data = data.replace(
    {'Oscar Year':
     {"1927/28":"1927", "1928/29":"1928", "1929/30":"1929", "1930/31":"1930",
"1931/32":"1931", "1932/33":"1932"
     }
    }
)
```

```python
data['Oscar Year'] = pd.to_numeric(data['Oscar Year'])
```

##### IMDB Votes

```python
data['IMDB Votes'] = data['IMDB Votes'].str.replace(',', '')
```

```python
data['IMDB Votes'] = pd.to_numeric(data['IMDB Votes'])
```

##### Original Release Date

```python
pd.to_datetime(data['Original Release Date'],format = '%Y-%m-%d')
```

##### Streaming Release Date

```python
pd.to_datetime(data['Streaming Release Date'],format = '%Y-%m-%d')
```

## 3.2. Пропущенные значения.

```python
data = data.dropna(subset = 'Tomatometer Rating')
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data = data.dropna(subset = 'Audience Status')
data = data.dropna(subset = 'Streaming Release Date')
data = data.dropna(subset = 'Directors')
data = data.dropna(subset = 'Authors')
```

```python
data = data.fillna({
    'Critic Consensus':'unknown',
    'Movie Info':'unknown',
})
```

```python
data.isna().sum()
```

## 3.3. Столб `Unnamed: 0`

```python
data = data.rename(columns={'Unnamed: 0' : 'id'})
```

# 4. EDA.

## 4.1. Исследование основных параметров.

```python
data.describe()[['Year of Release', 'Oscar Year', 'IMDB Rating', 'IMDB Votes', 'Tomatometer Rating', 'Tomatometer Count', 'Audience Rating', 'Audience Count', 'Tomatometer Top Critics Count', 'Tomatometer Fresh Critics Count', 'Tomatometer Rotten Critics Count']]
```

```python
columns = ['Year of Release', 'Oscar Year', 'IMDB Rating', 'Tomatometer Rating', 'Audience Rating', 'Audience Count', 'Tomatometer Top Critics Count', 'Tomatometer Fresh Critics Count', 'Tomatometer Rotten Critics Count']
```

```python
for column in columns:
    data.hist(column)
```

Столбцы с выбросами: 'IMDB Rating', 'Tomatometer Rating', 'Audience Rating', 'Audience Count', 'Tomatometer Fresh Ctritics Count', 'Tomatometer Rotten Critics Count'

```python
outlier = data[['IMDB Rating', 'Tomatometer Rating', 'Audience Rating', 'Audience Count', 'Tomatometer Fresh Critics Count', 'Tomatometer Rotten Critics Count']]
outlier
```

```python
Q1 = outlier.quantile(0.25)
Q3 = outlier.quantile(0.75)
IQR = Q3-Q1
```

```python
data_filtered = outlier[~((outlier < (Q1 - 1.5 * IQR)) |(outlier > (Q3 + 1.5 * IQR))).any(axis=1)]
data_filtered
```

```python
index_list = list(data_filtered.index.values)
```

```python
data_filtered = data[data.index.isin(index_list)]
data_filtered.head()
```

```python
filtered_outlier = data_filtered[['IMDB Rating', 'Tomatometer Rating', 'Audience Rating', 'Audience Count', 'Tomatometer Fresh Critics Count', 'Tomatometer Rotten Critics Count']]

for col in filtered_outlier:
    data_filtered.hist(col, bins=100)
```

```python
print (data.hist('IMDB Rating', bins=100))
print (data_filtered.hist('IMDB Rating', bins=100))

print (data.hist('Tomatometer Rating', bins=100))
print (data_filtered.hist('Tomatometer Rating', bins=100))

print (data.hist('Audience Rating', bins=100))
print (data_filtered.hist('Audience Rating', bins=100))

print (data.hist('Audience Count', bins=100))
print (data_filtered.hist('Audience Count', bins=100))

print (data.hist('Tomatometer Fresh Critics Count', bins=100))
print (data_filtered.hist('Tomatometer Fresh Critics Count', bins=100))

print (data.hist('Tomatometer Rotten Critics Count', bins=100))
print (data_filtered.hist('Tomatometer Rotten Critics Count', bins=100))
```

# 5. Заключение

## 5.1. Средняя оценка фильмов у каждой компании.

```python
average = data_filtered.pivot_table(index = 'Production Company', values = ['IMDB Rating', 'Tomatometer Rating', 'Audience Rating'], aggfunc='mean')
average
```

```python
imdb_stat = data_filtered.pivot_table(index='Production Company', values='IMDB Rating', aggfunc='mean')
imdb_stat = imdb_stat.sort_values(ascending=False, by = 'IMDB Rating')
imdb_stat = imdb_stat.reset_index()
imdb_stat.plot(grid=True, figsize=(20, 7), kind = 'bar', x = 'Production Company', y='IMDB Rating')
```

```python
aud_stat = data_filtered.pivot_table(index='Production Company', values='Audience Rating', aggfunc='mean')
aud_stat = aud_stat.sort_values(ascending=False, by = 'Audience Rating')
aud_stat = aud_stat.reset_index()
aud_stat.plot(grid=True, figsize=(20, 7), kind = 'bar', x = 'Production Company', y='Audience Rating')
```

```python
tomat_stat = data_filtered.pivot_table(index='Production Company', values='Tomatometer Rating', aggfunc='mean')
tomat_stat = tomat_stat.sort_values(ascending=False, by = 'Tomatometer Rating')
tomat_stat = tomat_stat.reset_index()
tomat_stat.plot(grid=True, figsize=(20, 7), kind = 'bar', x = 'Production Company', y='Tomatometer Rating')
```

**Заключение:**

- Студия, средняя оценка фильмов которой на IMDB является самой высокой - это Liberty Films.

- Студиии, средняя оценка фильмов которых на Rotten Tomatoes является самой высокой - это Image Entertainmetn Inc. и RKO.

- Студия, средняя оценка фильмов которой по мнению аудитории является самой высокой - это Liberty Films.

## 5.2. Зависимость оценки фильма от года выпуска. 

```python
imdb_corr = data_filtered.pivot_table(index='Year of Release', values='IMDB Rating', aggfunc='mean')
imdb_corr = imdb_corr.sort_values(ascending=True, by = 'Year of Release')
imdb_corr = imdb_corr.reset_index()
imdb_corr
```

```python
sns.lmplot(x="Year of Release", y="IMDB Rating", data=imdb_corr)
plt.title('IMDB to Year of Release Relation')
```

```python
imdb_c = imdb_corr[['Year of Release', 'IMDB Rating']]
corr_imdb = imdb_c.corr()
```

```python
round(corr_imdb, 3)
```

```python
tomat_corr = data_filtered.pivot_table(index='Year of Release', values='Tomatometer Rating', aggfunc='mean')
tomat_corr = tomat_corr.sort_values(ascending=True, by = 'Year of Release')
tomat_corr = tomat_corr.reset_index()
tomat_corr
```

```python
sns.lmplot(x="Year of Release", y="Tomatometer Rating", data=tomat_corr)
plt.title('Tomatometer to Year of Release Relation')
```

```python
tomat_c = tomat_corr[['Year of Release', 'Tomatometer Rating']]
corr_tomat = tomat_c.corr()
```

```python
round(corr_tomat, 3)
```

```python
aud_corr = data_filtered.pivot_table(index='Year of Release', values='Audience Rating', aggfunc='mean')
aud_corr = aud_corr.sort_values(ascending=True, by = 'Year of Release')
aud_corr = aud_corr.reset_index()
aud_corr
```

```python
sns.lmplot(x="Year of Release", y="Audience Rating", data=aud_corr)
plt.title('Audience to Year of Release Relation')
```

```python
aud_c = aud_corr[['Year of Release', 'Audience Rating']]
corr_aud = aud_c.corr()
```

```python
round(corr_aud, 3)
```

<br>
